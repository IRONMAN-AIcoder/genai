```python
from diffusers import StableDiffusionPipeline
from moviepy.editor import ImageSequenceClip
import torch
import os
from PIL import Image

# Load the Stable Diffusion 2.1 model
model_id = "stabilityai/stable-diffusion-2-1-base"  # Correct model identifier
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# Set the text prompt
prompt = "A dog walking in a park, realistic style"

# Create a folder to save frames
frames_dir = "frames"
os.makedirs(frames_dir, exist_ok=True)

# Generate 16 frames (8 fps * 2 seconds)
# Start with a random image as the base
init_image = pipe(prompt).images[0]

# Generate the first frame (initialize with the first image)
init_image.save(os.path.join(frames_dir, f"frame_000.png"))

# Generate subsequent frames using img2img for smooth transitions
for i in range(1, 16):
    # Use the last generated image as the base for the next frame
    image = pipe(prompt, init_image=init_image, strength=0.4).images[0]
    init_image = image  # Update the initial image for the next frame

    # Save the generated frame
    image.save(os.path.join(frames_dir, f"frame_{i:03d}.png"))

# Create the video using MoviePy
frame_files = [os.path.join(frames_dir, f) for f in sorted(os.listdir(frames_dir)) if f.endswith('.png')]
clip = ImageSequenceClip(frame_files, fps=8)

# Save the 2-second video
clip.write_videofile("generated_video.mp4", codec="libx264", audio=False)

# Display the generated video
from IPython.display import Video
Video("generated_video.mp4", embed=True)
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2"  # You can replace it with GPT-3 or GPT-4 if available
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Function to generate a story based on the context
def generate_story(context, max_length=500):
    # Encode the input context (the seed of the story)
    input_ids = tokenizer.encode(context, return_tensors='pt')

    # Generate the story using the model
    with torch.no_grad():
        output_ids = model.generate(input_ids, max_length=max_length, num_return_sequences=1, no_repeat_ngram_size=2, temperature=0.7)

    # Decode the output ids to a string
    story = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return story

# Example context (you can change this to any context you want)
context = """
Once upon a time, in a faraway land, there was a small village nestled between two tall mountains. The villagers lived simple but peaceful lives, and they had a strong bond with nature. However, one day, a mysterious stranger arrived, claiming to have knowledge of a hidden treasure. He said that only the bravest of the villagers could embark on a quest to find it.
"""

# Generate and print the story
story = generate_story(context)
print(story)
```
