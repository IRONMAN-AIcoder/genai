```python
import openai
import lamini

# Set API keys
OPENROUTER_API_KEY = "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"
LAMINI_API_KEY = "9c82623b9dce8bce5506acb332dfaaffb549e53405bf1f607b1c1f9e90943a49"

# Configure OpenRouter (OpenAI-style)
openai.api_key = OPENROUTER_API_KEY
openai.api_base = "https://openrouter.ai/api/v1"

# Configure Lamini
lamini.api_key = LAMINI_API_KEY
lamini_client = lamini.Lamini()

# Ask OpenRouter (Hugging Face model)
def ask_openrouter(prompt: str) -> str:
    try:
        response = openai.ChatCompletion.create(
            model="huggingface/mistralai/Mistral-7B-Instruct-v0.1",  # You can change this
            messages=[{"role": "user", "content": prompt}]
        )
        return response['choices'][0]['message']['content']
    except Exception as e:
        return f"[OpenRouter Error] {str(e)}"

# Ask Lamini
def ask_lamini(prompt: str) -> str:
    try:
        return lamini_client.generate(prompt)
    except Exception as e:
        return f"[Lamini Error] {str(e)}"

# Main CLI loop
def main():
    print("ğŸ¤– Welcome to the CLI Chatbot!")
    print("Choose your model: [openrouter / lamini]")

    while True:
        model = input("Model> ").strip().lower()
        if model not in ["openrouter", "lamini"]:
            print("Invalid model. Type 'openrouter' or 'lamini'.")
            continue

        while True:
            user_input = input("You> ")
            if user_input.lower() in ["exit", "quit"]:
                print("Goodbye! ğŸ‘‹")
                return

            if model == "openrouter":
                reply = ask_openrouter(user_input)
            else:
                reply = ask_lamini(user_input)

            print(f"{model.capitalize()}> {reply}\n")

if __name__ == "__main__":
    main()
```

```python
# âœ… Install the required libraries


# âœ… Imports
import whisper
from transformers import pipeline
import os
import subprocess
import uuid

# âœ… Function to download audio using yt-dlp
def download_audio(video_url, output_path="temp_audio.mp3"):
    try:
        # Generate unique filename to avoid conflicts
        unique_filename = f"yt_audio_{uuid.uuid4().hex}.mp3"
        output_template = f"{unique_filename}"

        command = [
            "yt-dlp",
            "-x", "--audio-format", "mp3",
            "-o", output_template,
            video_url
        ]
        subprocess.run(command, check=True)
        return output_template
    except subprocess.CalledProcessError as e:
        return f"[Error] Audio download failed: {str(e)}"

# âœ… Whisper transcription
def transcribe_audio_whisper(audio_path):
    try:
        model = whisper.load_model("base")
        result = model.transcribe(audio_path)
        os.remove(audio_path)
        return result["text"]
    except Exception as e:
        return f"[Error] Whisper failed: {str(e)}"

# âœ… Summarization
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_text(text):
    if text.startswith("[Error]"):
        return text
    max_chunk = 1000
    chunks = [text[i:i + max_chunk] for i in range(0, len(text), max_chunk)]
    summary = ""
    for chunk in chunks:
        summary += summarizer(chunk, max_length=130, min_length=30, do_sample=False)[0]['summary_text'] + " "
    return summary.strip()

# âœ… Run the summarizer bot
def run_summary_bot():
    print("ğŸ§ YouTube Video Summarizer (Whisper + yt-dlp)")
    print("Paste a video URL (or type 'exit' to quit)")
    while True:
        url = input("You> ")
        if url.lower() in ["exit", "quit"]:
            print("ğŸ‘‹ Bye!")
            break

        print("â³ Downloading audio using yt-dlp...")
        audio_file = download_audio(url)
        if "[Error]" in audio_file:
            print(f"\nğŸ“‹ Summary:\n{audio_file}\n")
            continue

        print("ğŸ™ï¸ Transcribing with Whisper...")
        transcript = transcribe_audio_whisper(audio_file)

        print("ğŸ§  Summarizing...")
        summary = summarize_text(transcript)
        print(f"\nğŸ“‹ Summary:\n{summary}\n")

run_summary_bot()
```
